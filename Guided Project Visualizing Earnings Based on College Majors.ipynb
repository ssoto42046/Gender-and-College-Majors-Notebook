# Exploring Gender and College Majors through Visualization
### The goal of this exploration is to answer the following questions:
* *Do students in more popular majors make more money?*
* _How many majors are predominantly male? Predonminantly female?_
* _Which category of majors have hte most students?_

We seek to answer these questions using basic plotting and visualization techniques, such as line plots, box plots, bar charts etc. We use the pandas and matplotlib libraries to achieve this.

### About our Dataset
Our dataset is on job outomces of students who finished college between 2010 and 2012. The dataset was released by American Commmunity Survey, and FiveThirtyEight cleaned the dataset and posted it on their Github repository. 

Our dataset contains the following columns and their meanings:

| Variable | Description |
| ----------- | ----------- |
|Rank | Rank by median earnings (the dataset is ordered by this column)|
| Major_code | Major code|
| Major | Major description |
| Major_category | Category of major |
| Total | Total number of people with major |
| Sample_size | Sample size (unweighted) of full-time |
| Men | Male graduates |
| Women | Female graduates |
| ShareWomen | Women as share of total |
| Employed | Number employed |
| Median | Median salary of full-time, year-round workers |
| Low_wage_jobs | Number in low-wage service jobs |
| Full_time | Number employed 35 hours or more |
| Part_time | Number employed less than 35 hours |

### Answering some questions:

_Do students in more popular majors make more money?_

There is a small bias that more popular majors have a smaller median salary. However that it is really hard to tell at this point since there are many similar majors that are overlapping, and are not compartmentalized into more distinct groups.

_Do students that majored in subjects that were majority female make more money?_

Based on the intial scatter plot, there looks that there is not really any relation between these two factors. The scatter plot is very sparse and there is not linear or any other relation between the two. 

_Is there any link between the number of full-time employees and median salary?_

This one is similar to the above question, but it is hard to say clearly.
